```python
spark.sql("""
CREATE SCHEMA IF NOT EXISTS workspace.ecommerce
""")
```

```python
spark.sql("CREATE VOLUME IF NOT EXISTS workspace.ecommerce.ecommerce_data")
```

```python
raw_path = "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv"
table_name = "workspace.ecommerce.ecommerce_delta"

# CSV Read
df_raw = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(raw_path)
)

# Managed Table
df_raw.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(table_name)
```

```python
# Append 1
df_raw.write \
    .format("delta") \
    .mode("append") \
    .saveAsTable(table_name)

# Append 2
df_raw.write \
    .format("delta") \
    .mode("append") \
    .saveAsTable(table_name)
```

```python
# History check
display(spark.sql(f"DESCRIBE HISTORY {table_name}"))

# OPTIMIZE
spark.sql(f"OPTIMIZE {table_name}")

# OPTIMIZED History
display(spark.sql(f"DESCRIBE HISTORY {table_name}"))
```
