```python
import time
from pyspark.sql.functions import col

start_time = time.time()

df = spark.table("workspace.ecommerce.ecommerce_delta")

total_rows = df.count()
print(f"Total Rows Scan 1: {total_rows}")

electronics_df = df.filter(col("category_code").contains("electronics"))

elec_rows = electronics_df.count()
print(f"Electronics Rows Scan 2: {elec_rows}")

results_list = electronics_df.groupBy("brand").count().collect()

for row in results_list[:5]:
    print(f"{row['brand']}: {row['count']}")

end_time = time.time()
time_expensive = end_time - start_time
print(f"\nTime Taken (Wallet Burner): {time_expensive:.2f} seconds")
```

```python
start_time_opt = time.time()

df = spark.table("workspace.ecommerce.ecommerce_delta")

optimized_df = (
    df.filter(col("category_code").contains("electronics"))
    .groupBy("brand")
    .count()
    .orderBy(col("count").desc())
)

display(optimized_df.limit(5))

end_time_opt = time.time()
time_optimized = end_time_opt - start_time_opt

print(f"\nTime Taken (Money Saver): {time_optimized:.2f} seconds")
```
