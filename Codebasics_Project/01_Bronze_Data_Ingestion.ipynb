```python
from pyspark.sql.functions import current_timestamp, lit

raw_csv_path = "/Volumes/workspace/aml_fraud_project/raw_data/paysim_raw.csv"
bronze_table_name = "workspace.aml_fraud_project.transactions_bronze"
```

```python
raw_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(raw_csv_path)

bronze_df = raw_df \
    .withColumn("ingestion_timestamp", current_timestamp()) \
    .withColumn("source_system", lit("kaggle_paysim"))
```

```python
print(f"Saving data to Delta Table: {bronze_table_name}")
bronze_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(bronze_table_name)

row_count = spark.table(bronze_table_name).count()
print(f"Bronze table created with {row_count} rows.")
display(spark.table(bronze_table_name).limit(5))
```
